```python
# Import pustaka yang diperlukan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()
```

```python
# membaca dataset
dataset = pd.read_csv('CADdataset.csv')
```

```python
# Tampilkan beberapa baris pertama dari dataset
dataset.head()
```

```python
dataset.info()
```

```python
# Membandingkan kolom yang bertipe data categorical dan numerik
cat_col = [col for col in dataset.columns if dataset[col].dtype == object]
print ("Categorical Kolom = ", cat_col)
num_col = [col for col in dataset.columns if dataset[col].dtype != object]
print ("Numeric Kolom = ", num_col)
```

```python
# Data Transformation
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

dataset['Sex'] = enc.fit_transform(dataset['Sex'].values)
dataset['Obesity'] = enc.fit_transform(dataset['Obesity'].values)
dataset['DLP'] = enc.fit_transform(dataset['DLP'].values)
dataset['WPE'] = enc.fit_transform(dataset['WPE'].values)
dataset['Atypical'] = enc.fit_transform(dataset['Atypical'].values)
dataset['Nonanginal'] = enc.fit_transform(dataset['Nonanginal'].values)
dataset['LVH'] = enc.fit_transform(dataset['LVH'].values)
dataset['PRP'] = enc.fit_transform(dataset['PRP'].values)
dataset['CAD'] = enc.fit_transform(dataset['CAD'].values)
```

```python
# Informasi Data Setelah diedit, dan diambil data yang dibutuhkan
dataset.info()
```

```python
# Menghapus data yang tidak akan digunakan atau bertipe data "Categorical", karena dalam pengolahan data machine learning harus menggunakan tipe data "Numeric"
dataset.drop(['Sex', 'Obesity', 'DLP', 'WPE', 'Atypical', 'Nonanginal', 'LVH', 'PRP'], axis=1, inplace=True) #Menyisakan satu kolom CAD karena akan menjadi target
```

```python
dataset.head()
```

```python
# Pembagian data X dan Y
x = dataset.iloc[:, 0:-1].values #masukkan data Age sampai BUN kedalam variabel x
y = dataset.iloc[:, -1].values   #masukkan data CAD kedalam variabel y
print(x)
print(y)
```

```python
y = dataset['CAD']
print(y)
```

```python
# Pembagian data Training dan Testing menggunakan library dari sklearn dengan function train_test_split
from sklearn.model_selection import train_test_split    #import library sklearn dan menggunakan function train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state=0)    #Membagi data training 75% dan data testing (0.25) 25%
```

```python
# Data preprocessing dengan StandardScaler
from sklearn.preprocessing import StandardScaler    #import library sklearn dan menggunakan function StandardScaler
sc = StandardScaler()   #Memasukkan fungsi StandardScaler kedalam variabel sc

x_train = sc.fit_transform(x_train)   #Scaling untuk x_train
x_test = sc.fit_transform(x_test)     #Scaling skala untuk x_test
```

```python
# Menggunakan metode
from sklearn.neighbors import KNeighborsClassifier
classifire = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski', p=2)   #Menentukan k=5 (n_neighbors=5) dengan perhitungan jarak menggunakan metode minkowski

classifire.fit(x_train, y_train)    #Mengklasifikasikan data x_train dan y_train
```

```python
# Menentukan prediksi
y_pred = classifire.predict(x_test)   #Menentukan nilai prediksi pada label/target dengan variabel y_pred, dan data x_test digunakan untuk memprediksi
```

```python
# Mendapatkan nilai akurasi dengan Evaluasi dan Validasi
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)   #Membandingkan nilai label/target sebenarnya (y_test) dengan target prediksi (y_pred)

print(cm)   #[[49  5]     49 + 11 : jumlah data(76) x 100
            #  [ 11 11]]   60 : 76 x  100 = 78,9% (Nilai Akurasi)
```
