```python
# Menghapus atribut car_ID karena tidak digunakan dalam klasifikasi
dataset.drop(['car_ID'], axis=1, inplace=True)
```

```python
# Melihat dimensi data
dataset.shape
```

**Preprocessing Data**

```python
#Data Transformasi
# Import library
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

# Konversi data dari categorical ke numeric
dataset['CarName'] = enc.fit_transform(dataset['CarName'].values)
dataset['fueltype'] = enc.fit_transform(dataset['fueltype'].values)
dataset['aspiration'] = enc.fit_transform(dataset['aspiration'].values)
dataset['doornumber'] = enc.fit_transform(dataset['doornumber'].values)
dataset['carbody'] = enc.fit_transform(dataset['carbody'].values)
dataset['drivewheel'] = enc.fit_transform(dataset['drivewheel'].values)
dataset['enginelocation'] = enc.fit_transform(dataset['enginelocation'].values)
dataset['enginetype'] = enc.fit_transform(dataset['enginetype'].values)
dataset['cylindernumber'] = enc.fit_transform(dataset['cylindernumber'].values)
dataset['fuelsystem'] = enc.fit_transform(dataset['fuelsystem'].values)
```

```python
# Memanggil 10 data teratas untuk memastikan bahwa konversi berhasil
dataset.head(11)
```

```python
# Menentukan nilai X yang akan digunakan sebagai klasifikasi
# Menggunakan semua atribut kecuali 'citympg' karena akan dijadikan sebagai target atau class
atr_data = dataset.drop(columns='citympg')
atr_data.head()
```

```python
# Merubah tipe data float ke int dan menentukan atribut 'citympg' sebagai class
cls_data = dataset['citympg'].astype(int)
cls_data.head(3)
```

```python
# Masukkan library dan masuk ke Algoritma Dicision Tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
# Membagi data training dan testing menggunakan function train_test_split
x_train, x_test, y_train, y_test = train_test_split(atr_data, cls_data, test_size = 0.2, random_state = 1)
tree_data = DecisionTreeClassifier (random_state = 1)
tree_data.fit(x_train, y_train)
```

```python
# Melihat akurasi dari hasil data testing
print("Nilai akurasi pada data testing : ", tree_data.score(x_test, y_test))
```

```python
# Import library dan membuat pohon keputusan(Dicision Tree)
from sklearn.tree import export_graphviz
export_graphviz(tree_data, out_file="tree_price.dot", feature_names=atr_data.columns, impurity=False, filled=True)
```
